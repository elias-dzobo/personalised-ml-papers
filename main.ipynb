```python
import requests as re 

url = 'https://paperswithcode.com'
```

```python
response = re.get(url)

response.content
```

```python
from bs4 import BeautifulSoup 


soup = BeautifulSoup(response.content, 'html5lib')
```

```python
print(soup.prettify())
```

```python
table = soup.findAll('div', attrs={'class': 'col-lg-9 item-content'})

print(table)
```

```python
print(table[1])
```

```python
print(table[1].h1.a.text)
print(table[1].h1.a['href'])
print(table[1].p.span.a['href'])
spans_without_class = [span for span in table[1].findAll('span') if not span.has_attr('class')]
for tag in spans_without_class:
    print(tag.text)
```

```python
new_url = url + table[1].h1.a['href']

res = re.get(new_url)

res.content
```

```python
new_soup = BeautifulSoup(res.content, 'html5lib')


print(new_soup.prettify())
```

```python
new_url
```

```python
new_table = new_soup.findAll('div', attrs={'class': 'paper-title'})

print(new_table)
```

```python
new_table[0]
```

```python
print(new_table[0].h1.text.strip())
for author in new_table[0].findAll('span', attrs={'class': 'author-span'}):
    print(author.text)
```

```python
another_table = new_soup.findAll('div', attrs={'class': 'paper-abstract'})

print(another_table)
```

```python
print(another_table[0].p.text.strip())
print(another_table[0].a['href'])
```
